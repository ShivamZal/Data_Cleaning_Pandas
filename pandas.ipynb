```python
# loading libraries

import pandas as pd
import numpy as np
```

```python
# reading dataframe

df = pd.read_csv("unclean_data.csv")
```

```python
# initial EDA

df
```

```python
# removing extra characters from title

df['movie_title'] = df['movie_title'].str[:-2]
```

```python
df
```

```python
df.duplicated('movie_title')
```

```python
df = df.drop_duplicates('movie_title')
```

```python
# changing na values in director likes

df['DIRECTOR_facebook_likes'] = pd.to_numeric(df['DIRECTOR_facebook_likes'].str.replace('"', ''))
df['DIRECTOR_facebook_likes'] = df['DIRECTOR_facebook_likes'].fillna(0)
```

```python
df
```

```python
# Fill it with the mean
df['duration'].mean()
```

```python
df['duration'] = df['duration'].fillna(df['duration'].mean())
```

```python
df
```

```python
# Calculate the 'hit' column based on the condition 'gross - budget > 0'
df['hit'] = df.apply(lambda row: 'yes' if row['gross'] - row['budget'] > 0 else 'no', axis=1)
```

```python
df
```

```python
df['gross'] = df['gross'] / 1000000
```

```python
df.rename(columns={'gross': 'gross (in millions'}, inplace=True)
```

```python
# Rename the 'budget' column to 'budget (in millions)'
df.rename(columns={'budget': 'budget (in millions)'}, inplace=True)

```

```python
# Divide the values in the 'budget (in millions)' column by 1,000,000
df['budget (in millions)'] = df['budget (in millions)'] / 1000000
```

```python
df = df.drop(columns=['title_year.1'])
```

```python
df
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df['ACTOR_2_facebook_likes'] = df['ACTOR_2_facebook_likes'].fillna(0)
```

```python
# List of columns to sum
columns_to_sum = ['DIRECTOR_facebook_likes', 'actor_3_facebook_likes', 'ACTOR_1_facebook_likes', 'ACTOR_2_facebook_likes']

# Calculate the sum of rows from the specified columns and assign to a new column 'Total_likes'
df['Total_likes'] = df[columns_to_sum].sum(axis=1)

# Display the modified DataFrame
print(df)
```

```python
df = df.drop(columns=['Cast_Total_facebook_likes'])
```

```python
df.isnull().sum()
```

```python
df['num_voted_users'] = df['num_voted_users'].fillna(df['num_voted_users'].mean())
```

```python
df['facenumber_in_poster'] = df['facenumber_in_poster'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.upper()
```

```python
# Save the DataFrame as a CSV file
df.to_csv('clean_data.csv', index=False)
```
